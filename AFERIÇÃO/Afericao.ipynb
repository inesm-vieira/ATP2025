a) contaEspacos: defina uma função que recebe uma string e devolve o número de espaços na string.

```python
def contaEspacos(s):
    if s == "":
        return 0
    if s[0] == " ":
        return 1 + contaEspacos(s[1:])
    else:
        return contaEspacos(s[1:])

print(contaEspacos("Olá chamo me Pedro!"))
```

b) ordenada?: defina uma função que recebe uma lista como argumento e devolve o valor booleano verdadeiro se a lista estiver ordenada e falso caso contrário.

```python
def ordenada(lista):
    if len(lista) <= 1:
        return True
    if lista[0] <= lista[1]:
        return ordenada(lista[1:])
    else:
        return False

print(ordenada([1,2,3,6,12]))
print(ordenada([1,1,3,6,12]))
print(ordenada([1,2,3,1,12]))
```

c) existeElem: defina uma função que recebe uma lista e um elemento e devolve True ou False caso o elemento exista ou não na lista.

```python
def existeElem(lista, elem):
    if lista == []:
        return False
    if lista[0] == elem:
        return True
    else:
        return existeElem(lista[1:], elem)

print(existeElem([1,2,3,4,5],7))
print(existeElem([1,2,3,4,5],2))
```

a) maiorDivisor: defina uma função que recebe um número  N, e calcula o maior número de um digito (1 a 9) que divide cada número de entre 1 e N. O resultado será um dicionário.

```python
def maiorDivisor(N):
    resultado = {}
    for n in range(1, N + 1):
        maior = 1
        for d in range(1, 10):
            if n % d == 0:
                maior = d
        resultado[n] = maior
    return resultado

print(maiorDivisor(20))
```

c) f2c: Converta um registo de temperaturas Fahrenheit, num novo registo em que a cada temperatura Fahrenheit associas uma temperatura em Celsius.

```python

fahrenheit = {'t1':-30, 't2':-20, 't3':-10, 't4':0}

def fah(conv):
    res = {}
    for chave in conv:
        f = conv[chave]
        c = (f - 32) / 1.8
        res[chave] = (f, c)
    return res

print(fah(fahrenheit))
```

d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

```python
def reduxInt(n):
    if n < 10:
        return n
    soma = 0
    for d in str(n):
        soma = soma + int(d)
    return reduxInt(soma)

print(reduxInt(998))
```

d) existe_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e verifica se o médico existe nessa unidade hospitalar, devolvendo  True caso exista e False em caso contrário.

```python
def existe_medico(hospital, n_unidade, n_medico):
    res = False
    for unidade in hospital:
        if unidade[0] == n_unidade:
            for medico in unidade[1]:
                if medico[1] == n_medico:
                    res = True
    return res

print(existe_medico(hospital, "Unidade de Cardiologia", "Dra. Maria"))
```

e) adicionar_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome de um médico, e caso este não exista na unidade, acrescenta-o e devolve o novo hospital com o médico inserido.

```python
def adicionar_medico(hospital, n_unidade, n_medico):
    novo_hospital = []

    for unidade in hospital:
        if unidade[0] == n_unidade:
            medicos = unidade[1][:]
            res = False

            for medico in medicos:
                if medico[1] == n_medico:
                    res = True

            if res == False:
                novo_id = len(medicos) + 1
                medicos.append((novo_id, n_medico))

            novo_hospital.append((unidade[0], medicos, unidade[2]))
        else:
            novo_hospital.append(unidade)

    return novo_hospital

print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra. Paula")) 
```

f) ocupar_cama: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o identificador de uma cama, e caso a cama esteja livre devolve um novo hospital com a cama ocupada (não faz nada se a cama estiver ocupada).

```python
def ocupar_cama(hospital, n_unidade, id_cama):
    novo_hospital = []

    for unidade in hospital:
        if unidade[0] == n_unidade:
            novas_camas = []

            for cama in unidade[2]:
                if cama[0] == id_cama and cama[1] == False:
                    novas_camas.append((cama[0], True))
                else:
                    novas_camas.append(cama)

            novo_hospital.append((unidade[0], unidade[1], novas_camas))
        else:
            novo_hospital.append(unidade)

    return novo_hospital

print(ocupar_cama(hospital, "Unidade de Cardiologia", 101)) 
```

j) unidades_ocupacao_critica: Defina uma função que recebe um hospital e devolve uma lista de tuplos (pares), em que o primeiro elemento do tuplo é o nome de uma unidade hospitalar e o segundo é a percentagem de camas ocupadas nessa unidade, mas apenas deverão constar tuplos em que a percentagem é maior ou igual a 80%.

```python
def unidades_ocupacao_critica(hospital):
    res = []

    for unidade in hospital:
        ocupadas = 0
        total = 0

        for cama in unidade[2]:
            total = total + 1
            if cama[1] == True:
                ocupadas = ocupadas + 1

        percentagem = (ocupadas / total) * 100

        if percentagem >= 80:
            res.append((unidade[0], percentagem))

    return res

print(unidades_ocupacao_critica(hospital))
```
